```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv("Divar.csv")
data.head(10)
```

```python
data.info()
```

```python
n_columns = [
    'cat2_slug', 'city_slug', 'created_at_month',      
    'price_value', 'transformable_price',                 
    'credit_value', 'rent_value',                         
    'transformable_credit', 'transformable_rent',         
    'location_latitude', 'location_longitude',           
    'construction_year', 'rooms_count', 'building_size'  
]

print(data[n_columns].info())
```

```python
target_categories = ['residential-sell', 'commercial-sell'] 
whole_sales_data = data[data['cat2_slug'].isin(target_categories)].copy()

whole_sales_data['created_at_month'] = pd.to_datetime(whole_sales_data['created_at_month'])

whole_sales_data['transformable_price'] = pd.to_numeric(whole_sales_data['transformable_price'], errors='coerce')
whole_sales_data['final_price'] = whole_sales_data['transformable_price'].fillna(whole_sales_data['price_value'])

whole_sales_data = whole_sales_data.dropna(subset=['final_price'])
whole_sales_data = whole_sales_data[whole_sales_data['final_price'] > 0]

monthly_market_price = whole_sales_data.groupby(pd.Grouper(key='created_at_month', freq='MS'))['final_price'].mean()
monthly_market_inflation = monthly_market_price.ffill().pct_change() * 100

annual_market_inflation = monthly_market_inflation.groupby(monthly_market_inflation.index.year).mean()

print(annual_market_inflation)
print(monthly_market_inflation)
print(monthly_market_price)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
monthly_market_inflation.plot(kind='line', marker='o', color='purple', title='Total Real Estate Market Inflation (%)')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--') 
plt.xlabel('Date')
plt.ylabel('Inflation Rate %')
plt.grid(True)
plt.show()
```

```python
import matplotlib.ticker as ticker
target_categories = ['residential-sell', 'commercial-sell']
df_market = data[data['cat2_slug'].isin(target_categories)].copy()
df_market['created_at_month'] = pd.to_datetime(df_market['created_at_month'])

df_market['transformable_price'] = pd.to_numeric(df_market['transformable_price'], errors='coerce')
df_market['final_price'] = df_market['transformable_price'].fillna(df_market['price_value'])
df_market = df_market[(df_market['final_price'] > 0) & (df_market['final_price'].notna())]

monthly_price = df_market.groupby(pd.Grouper(key='created_at_month', freq='MS'))['final_price'].mean()
monthly_inflation = monthly_price.ffill().pct_change() * 100

annual_price = df_market.groupby(df_market['created_at_month'].dt.year)['final_price'].mean()
annual_inflation = monthly_inflation.groupby(monthly_inflation.index.year).mean()

summary_table = pd.DataFrame({
    'Year': annual_price.index,
    'Avg Price (Billions)': annual_price.values / 1e9, 
    'Avg Monthly Inflation (%)': annual_inflation.values
})

print("\n Market Summary Table :")
print(summary_table.round(2).to_string(index=False))


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)
ax1.plot(monthly_price.index, monthly_price / 1e9, color='#1f77b4', linewidth=2, marker='o', markersize=4)
ax1.set_title('Whole Market Average Property Price (2021-2025)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Price (Billions of Tomans)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f} B'))

ax1.annotate('End of 2024 Spike\n(Luxury/Commercial Influx)', 
             xy=(pd.Timestamp('2024-11-01'), 35), 
             xytext=(pd.Timestamp('2023-01-01'), 45),
             arrowprops=dict(facecolor='red', shrink=0.05))

ax2.bar(monthly_inflation.index, monthly_inflation, color='gray', alpha=0.6, label='Monthly Inflation')

ax2.plot(monthly_inflation.index, monthly_inflation.rolling(3).mean(), color='red', linewidth=2, label='3-Month Trend')

ax2.set_title('Market Volatility (Monthly Inflation Rate)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Inflation (%)', fontsize=12)
ax2.set_xlabel('Date', fontsize=12)
ax2.axhline(0, color='black', linewidth=1)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
target_categories = ['residential-rent', 'commercial-rent']
df_rent = data[data['cat2_slug'].isin(target_categories)].copy()

df_rent['created_at_month'] = pd.to_datetime(df_rent['created_at_month'])

df_rent['final_price'] = pd.to_numeric(df_rent['transformable_credit'], errors='coerce')

df_rent = df_rent[(df_rent['final_price'] > 0) & (df_rent['final_price'].notna())]

monthly_rent_index = df_rent.groupby(pd.Grouper(key='created_at_month', freq='MS'))['final_price'].mean()

monthly_rent_inflation = monthly_rent_index.ffill().pct_change() * 100

annual_rent_avg = df_rent.groupby(df_rent['created_at_month'].dt.year)['final_price'].mean()
annual_rent_inflation = monthly_rent_inflation.groupby(monthly_rent_inflation.index.year).mean()

summary_table_rent = pd.DataFrame({
    'Year': annual_rent_avg.index,
    'Avg Full Mortgage (Millions)': annual_rent_avg.values / 1e6,
    'Avg Monthly Rent Inflation (%)': annual_rent_inflation.values
})

print("\n=== Rental Market Summary (Based on Full Mortgage Equivalent) ===")
print(summary_table_rent.round(2).to_string(index=False))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

ax1.plot(monthly_rent_index.index, monthly_rent_index / 1e6, color='#2ca02c', linewidth=2, marker='o', markersize=4)
ax1.set_title('Whole Rental Market Trends (Full Mortgage Equivalent)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Avg Mortgage (Millions of Tomans)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f} M'))

ax2.bar(monthly_rent_inflation.index, monthly_rent_inflation, color='gray', alpha=0.6, label='Monthly Inflation')
ax2.plot(monthly_rent_inflation.index, monthly_rent_inflation.rolling(3).mean(), color='green', linewidth=2, label='3-Month Trend')

ax2.set_title('Rental Market Volatility (Inflation)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Inflation (%)', fontsize=12)
ax2.set_xlabel('Date', fontsize=12)
ax2.axhline(0, color='black', linewidth=1)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
