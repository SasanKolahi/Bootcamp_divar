### Define Function For change Arabic and Farsi numeric Characters

```python
def persian_to_english_numbers(text):
    if not isinstance(text, str):
        return text

    persian_digits = '۰۱۲۳۴۵۶۷۸۹'
    arabic_digits  = '٠١٢٣٤٥٦٧٨٩'
    english_digits = '0123456789'

    translation_table = str.maketrans(
        persian_digits + arabic_digits,
        english_digits * 2
    )

    return text.translate(translation_table)
```

```python
data=data.applymap(persian_to_english_numbers)
```

### Define AdType Column and Add it to DataFrame

```python
def adType(row):
    if row in ['residential-rent','commercial-rent']:
        return 'rent'
    elif row in ['residential-sell','commercial-sell']:
        return 'sell'
    else:
        return 'RealStateService'
```

```python
data['ad_type']=data['cat2_slug'].apply(adType)
```

### Convert Created_at_month to dateTime

```python
#created_at_month to dateTime
data['created_at_month']=pd.to_datetime(data['created_at_month'],errors='coerce')
data['year_month'] = data['created_at_month'].dt.to_period('M')
```

### Good Format Plot

```python
df=data.copy()


# تبدیل rent_value به عدد 
df['transformable_rent'] = pd.to_numeric(df['transformable_rent'], errors='coerce')

# تبدیل created_at_month به ماه شمسی 
def greg_to_persian_month(date_str):
    try:
        dt = pd.to_datetime(date_str)
        jdt = jdatetime.date.fromgregorian(date=dt.date())
        return jdt.strftime('%B')  # e.g., 'Farvardin'
    except:
        return None

df['persian_month_en'] = df['created_at_month'].apply(greg_to_persian_month)

#  فیلتر آگهی‌های اجاره‌ای معتبر 
df_rent = df[
    (df['rent_value'] >= 0) &
    (df['cat2_slug'].str.contains('rent', case=False, na=False)) &
    (df['persian_month_en'].notna())
].copy()

print(f" تعداد آگهی‌های اجاره‌ای معتبر: {len(df_rent)}")

#  ترتیب ماه‌های شمسی 
persian_months_en = [
    'Farvardin', 'Ordibehesht', 'Khordad', 'Tir', 'Mordad', 'Shahrivar',
    'Mehr', 'Aban', 'Azar', 'Dey', 'Bahman', 'Esfand'
]
persian_months_fa = [
    'فروردین', 'اردیبهشت', 'خرداد', 'تیر', 'مرداد', 'شهریور',
    'مهر', 'آبان', 'آذر', 'دی', 'بهمن', 'اسفند'
]

month_map = dict(zip(persian_months_en, persian_months_fa))

# فیلتر بر اساس ماه‌های شناخته شده و تخصیص ترتیب
df_rent = df_rent[df_rent['persian_month_en'].isin(persian_months_en)]
df_rent['month_fa'] = df_rent['persian_month_en'].map(month_map)
df_rent['month_order'] = pd.Categorical(
    df_rent['persian_month_en'],
    categories=persian_months_en,
    ordered=True
).codes

#  محاسبه میانگین ماهانه 
monthly_stats = df_rent.groupby('month_fa').agg(
    mean_rent=('transformable_rent', 'mean'),
    count=('transformable_rent', 'count')
).reset_index()

# اضافه کردن ترتیب برای رسم صحیح
monthly_stats['month_order'] = monthly_stats['month_fa'].map({fa: i for i, fa in enumerate(persian_months_fa)})
monthly_stats = monthly_stats.sort_values('month_order')

# تبدیل به میلیون تومان
monthly_stats['mean_rent_mil'] = monthly_stats['mean_rent'] / 1_000_000

# رسم نمودار
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=monthly_stats['month_fa'],
    y=monthly_stats['mean_rent_mil'],
    mode='lines+markers+text',
    line=dict(color='#e74c3c', width=3, shape='spline'),
    marker=dict(size=10, color='#2c3e50', line=dict(width=2, color='#e74c3c')),
    text=[f"{x:.1f}" for x in monthly_stats['mean_rent_mil']],
    textposition="top center",
    textfont=dict(size=12, color='black'),
    name="میانگین اجاره (میلیون تومان)"
))


fig.update_layout(
    title=dict(
        text="ترند میانگین قیمت اجاره آگهی‌های املاک بر اساس ماه شمسی",
        font=dict(size=20, color='#2c3e50'),
        x=0.5
    ),
    xaxis=dict(
        title=dict(text="ماه شمسی", font=dict(size=16)), 
        tickfont=dict(size=14),  #  همچنان معتبر
        categoryorder='array',
        categoryarray=persian_months_fa,
        gridcolor='lightgray'
    ),
    yaxis=dict(
        title=dict(text="میانگین اجاره (میلیون تومان)", font=dict(size=16)),  
        tickfont=dict(size=14),
        gridcolor='lightgray'
    ),
    plot_bgcolor='white',
    hovermode="x unified",
    height=600,
    width=1000,
    margin=dict(l=40, r=40, t=80, b=60)
)

# اضافه کردن نوارهای شفاف برای حجم داده
fig.add_trace(go.Bar(
    x=monthly_stats['month_fa'],
    y=[max(monthly_stats['mean_rent_mil']) * 0.05] * len(monthly_stats),
    yaxis='y2',
    opacity=0.2,
    marker_color='gray',
    showlegend=False,
    hoverinfo='skip'
))

# محور دوم 
fig.update_layout(
    yaxis2=dict(
        overlaying='y',
        side='right',
        showgrid=False,
        visible=False
    )
)

# نمایش و ذخیره
fig.show()
fig.write_html("rent_trend_persian_months.html")
fig.write_image("rent_trend_persian_months.png", scale=2)
# ذخیره به صورت HTML (قابل باز کردن در مرورگر) و PNG
fig.write_html("rent_trend_persian_months.html")
# fig.write_image("rent_trend_persian_months.png", scale=2)

#  جدول خلاصه در خروجی 
print("\n خلاصه ترند میانگین اجاره (ماه شمسی):")
summary = monthly_stats[['month_fa', 'mean_rent_mil', 'count']].round({'mean_rent_mil': 1})
summary.columns = ['ماه شمسی', 'میانگین اجاره (میلیون تومان)', 'تعداد آگهی']
print(summary.to_string(index=False))
```

```python
# inflation calcualtion by cpi
df = data.copy()


#  تبدیل ستون‌های عددی 
df['price_value'] = pd.to_numeric(df['price_value'], errors='coerce')

#  استخراج سال شمسی از created_at_month 
def extract_shamsi_year(date_str):
    try:
        greg = pd.to_datetime(date_str)
        shamsi = jdatetime.date.fromgregorian(date=greg.date())
        return shamsi.year
    except:
        return None

df['shamsi_year'] = df['created_at_month'].apply(extract_shamsi_year)

# فیلتر آگهی‌های فروشی معتبر 
df_sell = df[
    (df['price_value'] > 0) &
    (df['cat2_slug'].str.contains('sell', case=False, na=False)) &
    (df['shamsi_year'].between(1400, 1403))
].copy()

print(f" آگهی‌های فروش معتبر در سال‌های ۱۴۰۰–۱۴۰۳: {len(df_sell)}")

# محاسبه میانگین قیمت اسمی به ازای هر سال 
nominal_stats = df_sell.groupby('shamsi_year')['price_value'].agg(
    mean_nominal='mean',
    count='count'
).reset_index()

# جدول CPI (مرکز آمار ایران) 
cpi_data = {
    1400: 393.4,
    1401: 450.1,
    1402: 510.3,
    1403: 570.9  # تخمین بر اساس روند + تورم 11.9% در 1402
}

nominal_stats['CPI'] = nominal_stats['shamsi_year'].map(cpi_data)
nominal_stats['real_price'] = (
    nominal_stats['mean_nominal'] * 393.4 / nominal_stats['CPI']
)

# تبدیل به میلیارد تومان برای خوانایی
nominal_stats['mean_nominal_bil'] = nominal_stats['mean_nominal'] / 1_000_000_000
nominal_stats['real_price_bil'] = nominal_stats['real_price'] / 1_000_000_000

# رسم نمودار مقایسه 
fig = go.Figure()

# قیمت اسمی (خط نارنجی)
fig.add_trace(go.Scatter(
    x=nominal_stats['shamsi_year'],
    y=nominal_stats['mean_nominal_bil'],
    mode='lines+markers+text',
    name='قیمت اسمی (میلیارد تومان)',
    line=dict(color='#e67e22', width=3),
    marker=dict(size=10, symbol='circle'),
    text=[f"{x:.1f}" for x in nominal_stats['mean_nominal_bil']],
    textposition="top center"
))

# قیمت حقیقی (خط آبی)
fig.add_trace(go.Scatter(
    x=nominal_stats['shamsi_year'],
    y=nominal_stats['real_price_bil'],
    mode='lines+markers+text',
    name='قیمت حقیقی (میلیارد تومان)',
    line=dict(color='#3498db', width=3, dash='dash'),
    marker=dict(size=10, symbol='square'),
    text=[f"{x:.1f}" for x in nominal_stats['real_price_bil']],
    textposition="bottom center"
))

# تنظیمات ظاهری 
fig.update_layout(
    title=dict(
        text="مقایسه قیمت اسمی و قیمت حقیقی (تعدیل‌شده با تورم) آگهی‌های فروش — سال‌های ۱۴۰۰ تا ۱۴۰۳",
        font=dict(size=18),
        x=0.5
    ),
    xaxis=dict(
        title=dict(text="سال شمسی", font=dict(size=14)),
        tickmode='array',
        tickvals=[1400, 1401, 1402, 1403],
        ticktext=['۱۴۰۰', '۱۴۰۱', '۱۴۰۲', '۱۴۰۳'],
        tickfont=dict(size=13)
    ),
    yaxis=dict(
        title=dict(text="میانگین قیمت (میلیارد تومان)", font=dict(size=14)),
        tickfont=dict(size=13),
        range=[0, nominal_stats['mean_nominal_bil'].max() * 1.1]
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    plot_bgcolor='white',
    hovermode="x unified",
    width=900,
    height=550
)

fig.show()
fig.write_html("real_vs_nominal_price_1400_1403.html")

# نمایش جدول 
print("\n جدول مقایسه قیمت اسمی و حقیقی:")
summary = nominal_stats[['shamsi_year', 'count', 'mean_nominal_bil', 'real_price_bil']].round(1)
summary.columns = ['سال شمسی', 'تعداد آگهی', 'قیمت اسمی (میلیارد)', 'قیمت حقیقی (میلیارد)']
print(summary.to_string(index=False))

# تحلیل روند 
last_real = nominal_stats[nominal_stats['shamsi_year'] == 1403]['real_price_bil'].iloc[0]
first_real = nominal_stats[nominal_stats['shamsi_year'] == 1400]['real_price_bil'].iloc[0]
change_pct = (last_real - first_real) / first_real * 100

print(f"\n روند قیمت حقیقی (۱۴۰۰ → ۱۴۰۳): {change_pct:+.1f}%")
if change_pct > 2:
    trend = "روند صعودی واقعی (رشد ارزش فراتر از تورم)"
elif change_pct < -2:
    trend = "روند نزولی واقعی (کاهش ارزش فراتر از تورم)"
else:
    trend = "ثبات نسبی (رشد قیمت تقریباً برابر تورم)"

print(f" نتیجه‌گیری: {trend}")
```

## PartII-Question2

```python
df=data.copy()

#  تبدیل ستون‌های عددی 
df['building_size'] = pd.to_numeric(df['building_size'], errors='coerce')
df = df[(df['building_size'] > 10) & (df['building_size'] < 1000)]

# پاک‌سازی و استخراج سال ساخت 
def parse_construction_year(val):
    if pd.isna(val):
        return None
    s = str(val).strip()
    # جستجوی الگوی سال (عدد چهار رقمی یا قبل از ...)
    if 'قبل از' in s:
        return 1370  # فرض: "قبل از ۱۳۷۰" → 1370
    # استخراج اولین عدد چهار رقمی
    match = re.search(r'1[3-4]\d{2}', s)  # سال 1300 تا 1499
    if match:
        return int(match.group())
    return None

df['year_parsed'] = df['construction_year'].apply(parse_construction_year)

#  فیلتر آگهی‌های مسکونی + سال ساخت موجود 
df_res = df[
    (df['cat2_slug'].str.contains('residential', case=False, na=False)) &
    (df['year_parsed'].notna()) &
    (df['building_size'].notna())
].copy()

print(f" تعداد آگهی‌های مسکونی با سال ساخت مشخص: {len(df_res)}")

#  دسته‌بندی به قدیمی/جدید 
df_res['is_old'] = df_res['year_parsed'] < 1396
df_res['age_group'] = df_res['is_old'].map({True: 'قدیمی (قبل از ۱۳۹۶)', False: 'جدید (۱۳۹۶ به بعد)'})

#  محاسبه میانگین مساحت 
summary = df_res.groupby('age_group')['building_size'].agg(
    count='count',
    mean='mean',
    median='median',
    std='std'
).round(1)

print("\n خلاصه مساحت زیربنا:")
print(summary.to_string())

#  آزمون t-test 
old = df_res[df_res['is_old']]['building_size']
new = df_res[~df_res['is_old']]['building_size']

if len(old) > 1 and len(new) > 1:
    t_stat, p_val = stats.ttest_ind(old, new, equal_var=False)  #  t-test
    print(f"\n آزمون t-test: t = {t_stat:.3f}, p = {p_val:.4f}")
    if p_val < 0.05:
        print(" تفاوت معنی‌دار آماری دارد (p < 0.05)")
    else:
        print(" تفاوت معنی‌دار آماری ندارد (p ≥ 0.05)")

#  رسم Boxplot با Plotly 
fig = px.box(
    df_res,
    x='age_group',
    y='building_size',
    color='age_group',
    color_discrete_map={'قدیمی (قبل از ۱۳۹۶)': '#2ecc71', 'جدید (۱۳۹۶ به بعد)': '#e74c3c'},
    category_orders={"age_group": ['قدیمی (قبل از ۱۳۹۶)', 'جدید (۱۳۹۶ به بعد)']},
    labels={'age_group': 'گروه سنی ملک', 'building_size': 'مساحت زیربنا (متر مربع)'},
    title='مقایسه مساحت خانه‌های قدیمی‌ساخت و جدیدساخت',
    points=False
)

# افزودن خط میانگین
means = df_res.groupby('age_group')['building_size'].mean()
for i, (group, mean_val) in enumerate(means.items()):
    fig.add_hline(
        y=mean_val,
        line_dash="dot",
        line_color="black",
        annotation_text=f"میانگین: {mean_val:.0f} متر²",
        annotation_position="top right" if i == 0 else "top left"
    )

fig.update_layout(
    showlegend=False,
    height=550,
    title_font_size=18
)
fig.write_html("old_vs_new_building_size.html")
fig.show()
```

## PartII-Question3

```python

df=data.copy()
#  فیلتر: فقط ملک‌های تجاری فروشی با قیمت معتبر 
df_com = df[
    (df['cat2_slug'].str.contains('commercial', case=False, na=False)) &
    (df['price_value'].notna())
].copy()

# تبدیل price_value به عددی
df_com['price_value'] = pd.to_numeric(df_com['price_value'], errors='coerce')
df_com = df_com[df_com['price_value'] > 0]


#  تبدیل has_business_deed به boolean 
def to_bool_business_deed(val):
    if pd.isna(val):
        return False
    s = str(val).strip().lower()
    return s in ['true', '1', 'yes']

df_com['has_business_deed_bool'] = df_com['has_business_deed'].apply(to_bool_business_deed)

# محاسبه میانگین قیمت بر اساس داشتن سند 
summary = df_com.groupby('has_business_deed_bool')['price_value'].agg(
    count='count',
    mean='mean',
    median='median',
    std='std'
).round()

summary.index = ['بدون سند تجاری', 'با سند تجاری']
summary = summary.reset_index()

print(" میانگین قیمت فروش ملک‌های تجاری (تومان):")
print(summary.to_string(index=False))

#  آزمون t-test (معنی‌داری آماری) 
with_deed = df_com[df_com['has_business_deed_bool'] == True]['price_value'].dropna()
without_deed = df_com[df_com['has_business_deed_bool'] == False]['price_value'].dropna()

if len(with_deed) > 1 and len(without_deed) > 1:
    t_stat, p_val = stats.ttest_ind(with_deed, without_deed, equal_var=False)  #  t-test
    print(f"\n آزمون t-test:")
    print(f"   t = {t_stat:.3f}, p = {p_val:.4f}")
    if p_val < 0.05:
        print(" تفاوت معنی‌دار آماری دارد (p < 0.05)")
        if with_deed.mean() > without_deed.mean():
            print(" ملک‌های با سند تجاری به‌طور معنی‌داری گران‌ترند.")
        else:
            print(" ملک‌های با سند تجاری ارزان‌ترند (کمتر رایج).")
    else:
        print(" تفاوت از نظر آماری معنی‌دار نیست (p ≥ 0.05)")
else:
    print("\n تعداد داده برای آزمون کافی نیست.")

#  رسم Boxplot مقایسه 
fig = px.box(
    df_com,
    x='has_business_deed_bool',
    y='price_value',
    color='has_business_deed_bool',
    color_discrete_map={False: '#e74c3c', True: '#2ecc71'},
    category_orders={"has_business_deed_bool": [False, True]},
    labels={
        'has_business_deed_bool': 'وضعیت سند تجاری',
        'price_value': 'قیمت فروش (تومان)'
    },
    title='تأثیر سند تجاری بر قیمت فروش ملک‌های تجاری',
    points=False

)

# تغییر برچسب محور x
fig.update_xaxes(
    tickvals=[False, True],
    ticktext=['بدون سند تجاری', 'با سند تجاری']
)
fig.update_yaxes(type="log", title="قیمت فروش (تومان — مقیاس لگاریتمی)")
# اضافه کردن خط میانگین
means = df_com.groupby('has_business_deed_bool')['price_value'].mean()
for i, (deed, mean_val) in enumerate(means.items()):
    fig.add_hline(
        y=mean_val,
        line_dash="dot",
        line_color="black",
        annotation_text=f"میانگین: {mean_val/1e6:.1f}M تومان",
        annotation_position="top left" if i == 0 else "top right"
    )

fig.update_layout(
    showlegend=False,
    height=550,
    yaxis_tickformat='.0s',
    title_font_size=18
)
fig.write_html("business_deed_impact.html")
fig.show()
```

## PartII-Question4

```python
df = data.copy()


#  تبدیل ستون‌های boolean به True/False 
def to_bool(val):
    if pd.isna(val):
        return False
    s = str(val).strip().lower()
    return s in ['true', '1', 'yes', 'on']

luxury_cols = ['has_pool', 'has_barbecue', 'has_sauna', 'has_jacuzzi']
basic_cols = ['has_balcony', 'has_elevator', 'has_warehouse', 'has_parking', 'has_security_guard']

for col in luxury_cols + basic_cols:
    df[col] = df[col].apply(to_bool)

#  فیلتر: فقط آگهی‌های فروشی با قیمت معتبر 
df_sell = df[
    (df['cat2_slug'].str.contains('sell', case=False, na=False)) &
    (df['price_value'].notna())
].copy()

df_sell['price_value'] = pd.to_numeric(df_sell['price_value'], errors='coerce')
df_sell = df_sell[df_sell['price_value'] > 0]

#  تعریف گروه‌ها 
df_sell['has_luxury'] = df_sell[luxury_cols].any(axis=1)
df_sell['has_basic'] = df_sell[basic_cols].any(axis=1)

print(f"✅ تعداد آگهی‌های فروشی معتبر: {len(df_sell)}")
print(f"   - دارای لوکس: {df_sell['has_luxury'].sum()}")
print(f"   - دارای پایه: {df_sell['has_basic'].sum()}")
print(f"   - دارای هر دو: {(df_sell['has_luxury'] & df_sell['has_basic']).sum()}")

#  محاسبه میانگین قیمت برای چهار گروه کلیدی 
# الف) دارای لوکس (صرف‌نظر از پایه)
luxury_yes = df_sell[df_sell['has_luxury']]['price_value']
luxury_no = df_sell[~df_sell['has_luxury']]['price_value']

# ب) دارای پایه (صرف‌نظر از لوکس)
basic_yes = df_sell[df_sell['has_basic']]['price_value']
basic_no = df_sell[~df_sell['has_basic']]['price_value']

# محاسبه خلاصه‌ها
summary = pd.DataFrame({
    'گروه': [
        'بدون لوکس', 'با لوکس',
        'بدون امکانات پایه', 'با امکانات پایه'
    ],
    'تعداد': [
        len(luxury_no), len(luxury_yes),
        len(basic_no), len(basic_yes)
    ],
    'میانگین قیمت (میلیارد تومان)': [
        luxury_no.mean() / 1e9,
        luxury_yes.mean() / 1e9,
        basic_no.mean() / 1e9,
        basic_yes.mean() / 1e9
    ]
}).round(2)

print("\n📊 میانگین قیمت فروش بر اساس داشتن امکانات:")
print(summary.to_string(index=False))

#  آزمون‌های t-test 
print("\n🔍 آزمون‌های آماری (t-test):")

# الف) لوکس
if len(luxury_yes) > 1 and len(luxury_no) > 1:
    t1, p1 = stats.ttest_ind(luxury_yes, luxury_no, equal_var=False)
    diff_pct_luxury = (luxury_yes.mean() - luxury_no.mean()) / luxury_no.mean() * 100
    print(f"• لوکس: t = {t1:.2f}, p = {p1:.4f} → {' معنی‌دار' if p1 < 0.05 else ' غیرمعنی‌دار'} | +{diff_pct_luxury:+.1f}%")

# ب) پایه
if len(basic_yes) > 1 and len(basic_no) > 1:
    t2, p2 = stats.ttest_ind(basic_yes, basic_no, equal_var=False)
    diff_pct_basic = (basic_yes.mean() - basic_no.mean()) / basic_no.mean() * 100
    print(f"• پایه: t = {t2:.2f}, p = {p2:.4f} → {' معنی‌دار' if p2 < 0.05 else ' غیرمعنی‌دار'} | +{diff_pct_basic:+.1f}%")

# رسم Barplot مقایسه 
data_plot = pd.DataFrame({
    'نوع امکانات': ['لوکس', 'لوکس', 'پایه', 'پایه'],
    'وضعیت': ['بدون', 'با', 'بدون', 'با'],
    'میانگین قیمت': [
        luxury_no.mean() / 1e9,
        luxury_yes.mean() / 1e9,
        basic_no.mean() / 1e9,
        basic_yes.mean() / 1e9
    ],
    'تعداد': [len(luxury_no), len(luxury_yes), len(basic_no), len(basic_yes)]
})

fig = px.bar(
    data_plot,
    x='نوع امکانات',
    y='میانگین قیمت',
    color='وضعیت',
    barmode='group',
    color_discrete_map={'بدون': '#e74c3c', 'با': '#2ecc71'},
    text='میانگین قیمت',
    title='مقایسه میانگین قیمت فروش با و بدون امکانات لوکس و پایه',
    labels={'میانگین قیمت': 'میانگین قیمت (میلیارد تومان)'}
)

fig.update_traces(texttemplate='%{text:.1f}', textposition='outside')
fig.update_layout(
    yaxis_range=[0, data_plot['میانگین قیمت'].max() * 1.15],
    height=500,
    title_font_size=18
)
fig.write_html("luxury_vs_basic_impact.html")
fig.show()
```
